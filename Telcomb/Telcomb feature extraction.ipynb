```python
import os
import pandas as pd
```

```python
### FOR AMR FEATURE EXTRACTION ###

# Path to the empty CSV file
arg_csv_file = "Telcomb_AMR_analytical_matrix.csv"

# Path to the directory containing the sample files
input_dir = "Telcomb_AMR_files"
```

```python
# Load the empty CSV file
arg_df = pd.read_csv(arg_csv_file, index_col=0)
```

```python
# Iterate through each sample file in the directory
for sample_file in os.listdir(input_dir):
    # Get the sample name from the file name
    sample_name = os.path.splitext(sample_file)[0].split("_")[0]
    
    # Read the sample file, starting at row 20
    sample_path = os.path.join(input_dir, sample_file)
    sample_data = pd.read_csv(sample_path, skiprows=19, usecols=[0, 1], header=None, names=["gene", "reads"])
    
    # Map the values from the sample file to the corresponding column in the DataFrame
    for index, row in sample_data.iterrows():
        gene = row["gene"]
        reads = row["reads"]
        if gene in arg_df.index:
            arg_df.loc[gene, sample_name] = reads
```

```python
# Save the updated DataFrame back to the CSV file
arg_df.to_csv(arg_csv_file)
```

```python
### FOR MGE FEATURE EXTRACTION ###

# Path to the empty CSV file
mge_csv_file = "Telcomb_MGE_analytical_matrix.csv"

# Path to the directory containing the sample files
input_dir = "Telcomb_MGE_files"
```

```python
# Load the empty CSV file
mge_df = pd.read_csv(mge_csv_file, index_col=0)
```

```python
# Iterate through each sample file in the directory
for sample_file in os.listdir(input_dir):
    # Get the sample name from the file name
    sample_name = os.path.splitext(sample_file)[0].split("_")[0]
    
    # Read the sample file, starting at row 21
    sample_path = os.path.join(input_dir, sample_file)
    sample_data = pd.read_csv(sample_path, skiprows=20, usecols=[0, 1], header=None, names=["gene", "reads"])
    
    # Ensure unique gene names across all files
    unique_genes = set(sample_data["gene"])  # Get unique gene names in the current file
    current_genes = set(mge_df.index)           # Get current genes in the DataFrame
    new_genes = unique_genes - current_genes  # Find new genes not yet in the DataFrame
    print(len(new_genes)," new genes added")
    
    # Add new genes as rows with NaN values for existing columns
    for new_gene in new_genes:
        mge_df.loc[new_gene] = [float("nan")] * len(mge_df.columns)
    
    # Map the values from the sample file to the corresponding column in the DataFrame
    for index, row in sample_data.iterrows():
        gene = row["gene"]
        reads = row["reads"]
        if gene in mge_df.index:
            mge_df.loc[gene, sample_name] = reads
    
```
